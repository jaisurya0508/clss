```python
numerical_data = data.select_dtypes(exclude='object')
numerical_data = numerical_data.dropna()
```

```python
def check_outliers(data_check):
    total_values = data_check.shape[0]
    print(total_values)
    df_out = pd.DataFrame(columns=['Selected Columns','No of Outliers',
                                   'Total No of Values', 'Outlier Pro'])
    for i in data_check.columns:
        q1 = np.percentile(data_check,25)
        q3 = np.percentile(data_check,75)
        iqr = q3-q1
        lower_fence = q1-(iqr*1.5)
        upper_fence = q3+(iqr*1.5)
        temp = data_check[(data_check[i] < lower_fence) | (data_check[i] > upper_fence)]
        n_outliers = temp.shape[0]
        outlier_prop = (n_outliers/total_values)*100
        df_out = df_out._append({'Selected Columns': i, 'No of Outliers':n_outliers,
                                'Total No of Values':total_values,   
                                'Outlier Pro':outlier_prop},ignore_index=True)
    df_out = df_out.sort_values(['No of Outliers'],ascending=False).reset_index(drop=True)
    return df_out
    
        
        

        
     
```

```python
check_outliers(numerical_data)
```

```python
def cap_outliers(data):
    for i in data.columns:
        q1 = np.percentile(data,25)
        q3 = np.percentile(data,75)
        iqr = q3-q1
        lower_bound = q1-(iqr*1.5)
        upper_bound = q3+(iqr*1.5) 
        data[i] = data[i].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
    return data
```

```python
cap_outliers(numerical_data)
```

```python
def check_outlier_columns(data, column):
    no_rows = data.shape[0]
    print(no_rows)
    df_out = pd.DataFrame(columns=['Selected Columns','No of Outliers','Total No of Values','Outlier Pro'])
    q1 = np.percentile(data,25)
    q3 = np.percentile(data,75)
    iqr = q3-q1
    lower = q1-(iqr*1.5)
    upper = q3 + (iqr*1.5)
    temp = data[(data[column] < lower ) | (data[column] > upper)]
    n_outliers = temp.shape[0]
    out_prop = (n_outliers/no_rows)*100
    df_out = df_out._append({'Selected Columns': column,
                            'No of Outliers':n_outliers,
                            'Total No of Values':no_rows,   
                            'Outlier Pro':out_prop},ignore_index=True)
    return df_out
    
    
```

```python
check_outlier_columns(numerical_data,'depression')
```
